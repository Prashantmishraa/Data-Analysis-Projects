### Removing outliers from Sales

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
IQR = Q3-Q1
threshold=1.5 
admit_outliers=df[(df['Sales'] <Q1-threshold*IQR) | (df['Sales'] > Q3+ threshold*IQR)]
admit_outliers
```

```python
 df=df.drop(admit_outliers.index)
 sns.boxplot(x='Sales',data=df)
```



### PLOTTING A GRAPH FOR CATEGORY-WISE SALES



### Number of orders country-wise 



```python
order_count = df.groupby('Country').count()['Row ID']
```

```python
order_count
```

```python
order_count.plot()
plt.xlabel("Country")
plt.ylabel("order_count")
#order_count.set_xlabel('x-axis', fontsize=12)
plt.title("Number of orders in Country")
plt.legend(bbox_to_anchor=(1.05,1))
plt.figure(figsize=(20,20))
```



### PLOTTING A GRAPH FOR MARKET-WISE SALES
